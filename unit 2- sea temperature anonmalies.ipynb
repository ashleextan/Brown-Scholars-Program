```python
import matplotlib.pyplot as plt
import numpy as np
import csv
import pandas as pd
```

Code for Dec 2002 Map

```python
csvFile = open("/content/drive/MyDrive/Grade 11.5/BridgeUP/Datasets/sea_temp_anom_Dec_2002.csv", 'r') 
csvReader = csv.reader(csvFile, delimiter=',')
data_lists = list(csvReader)
```

```python
print(data_lists[0])
len(data_lists[0])
len(data_lists)
```

```python
width = len(data_lists[0])
height = len(data_lists)
```

```python
grid = np.empty([height, width, 3], dtype=np.uint8)

all_values = [] 

for r, row in enumerate(data_lists): 
    for c, value in enumerate(row):  
        if float(value) < 99998:
            all_values.append(float(value)) 
        if float(value) <= -5:
            grid[r, c] = [46, 46, 255]
        elif float(value) <= -4:
            grid[r, c] = [97, 97, 255]
        elif float(value) <= -3:
            grid[r, c] = [145, 169, 255]
        elif float(value) <= -2:
            grid[r, c] = [199, 211, 255]
        elif float(value) <= -1:
            grid[r, c] = [230, 230, 255]
        elif float(value) <= 0:
            grid[r, c] = [245, 245, 245]
        elif float(value) <= 1:
            grid[r, c] = [255, 181, 181]
        elif float(value) <= 2:
            grid[r, c] = [255, 150, 150]
        elif float(value) <= 3:
            grid[r, c] = [255, 117, 117]
        elif float(value) <= 4:
            grid[r, c] = [255, 79, 79]
        elif float(value) <= 5:
            grid[r, c] = [255, 43, 43]
        else:
            grid[r, c] = [0, 0, 0]
```

```python
print(max(all_values))
print(min(all_values))
```

```python
plt.title("December 2002 Sea Temperature Anonmolies in Celsius")
plt.imshow(grid)
plt.axis('off')
plt.savefig("Sea_temperature_2002.png",dpi = 1000)
plt.show()
```

Code for Sea Temperature Line Graph

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/Grade 11.5/BridgeUP/Datasets/SeaTemp_SeaLevel_anomaly.csv")
```

```python
print(df.head())
len(df)
```

```python
years= df["Year"]
tempAnomaly = df['Sea temperature anomaly (C)']
```

```python
plt.plot(years, tempAnomaly)

plt.xlim(1992, 2016)
plt.xticks(range(1992, 2018, 4))
plt.title("Temperature Anonmalies From 1992-2016")
plt.xlabel("Years from 1992-2016")
plt.ylabel("Sea Temperature Anomalies (C)")

plt.show()
```
